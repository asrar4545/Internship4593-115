```python
train = d1.sample(frac=0.8,random_state=123)
test = d1.drop(train.index)
```

```python
print(train.shape)
print(test.shape)
```

```python
X_train=train.drop(['RetentionPotential'],axis=1)
y_train=train['RetentionPotential']
X_test=test.drop(['RetentionPotential'],axis=1)
y_test=test['RetentionPotential']
```

```python
X_train = X_train.values
y_train = y_train.values
X_test = X_test.values
y_test = y_test.values
```

```python
class KNN:
    def __init__(self, k=3):
        self.k = k
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        
    def predict(self, X_test):
        y_pred = []
        for row in X_test:
            dist = np.sqrt(np.sum((row - self.X_train)**2, axis=1))
            idx = np.argsort(dist)[:self.k]
            labels = self.y_train[idx]
            counts = np.bincount(labels)
            y_pred.append(np.argmax(counts))
        return y_pred
```

```python
def knn(X_train, y_train, X_test, k):
    knn = KNN(k=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    return y_pred
```

```python
accuracy = sum(y_pred == y_test)/len(y_test)
print(f"Accuracy: {accuracy:.2f}")
```
